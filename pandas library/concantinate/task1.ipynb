```python
import pandas as pd
india_weather=pd.DataFrame({
    "city":["mumbai","dehli","banglore"],
    "temperature":[32,45,90],
    "humidity":[80,70,60]
})
india_weather
```

```python
UK=pd.DataFrame({
    "city":["new york","chicago","orandlo"],
    "temperature":[34,56,89],
    "humidity":[812,123,45]
})
UK
```

```python
df=pd.concat([india_weather,UK])
df
```

```python
df=pd.concat([india_weather,UK],ignore_index=True)
df
```

```python
df=pd.concat([india_weather,UK],keys=["india","UK"])
df
```

```python
df.loc["UK"]
```

```python
df.loc["india"]
```

```python
Temperature=pd.DataFrame({
    "city":["mumbai","dehli","banglore"],
    "temperature":[32,45,90],
    "humidity":[80,70,60]
},index=[0,1,2])
Temperature
```

```python
windspeed=pd.DataFrame({
    "city":["dehli","mumbai"],
    "windspeed":[7,12]
},index=[1,0])
windspeed
```

```python
df=pd.concat([Temperature,windspeed],axis=1)
df
```

```python
s=pd.Series(["Humid","Dry","Rain"],name="event")
s
```

```python
df=pd.concat([Temperature,s],axis=1)
df
```
