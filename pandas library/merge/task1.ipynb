```python
import pandas as pd

df1=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35]
})
df1
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","orlando"],
    "humidity":[65,45,75]
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df1=pd.DataFrame({
    "city":["new york","chicago","multan","karachi"],
    "temperature":[21,14,35,89]
})
df1
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","fsd","layyah"],
    "humidity":[65,45,75,90]
})
df2
```

```python
df3=pd.merge(df1,df2,on="city") # merge ma sirf common city ay gy
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer") #is ma dono city ka union ay ga
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left") # how left ma left walu element aur common elements dono ay gy
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right") # how right ma right walu element aur common elements dono ay gy
df3
```

```python
df1=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[65,45,75]
})
df1
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","orlando"],
    "humidity":[65,45,75],
    "temperature":[21,14,35],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df3=pd.merge(df1,df2,on="city",suffixes=("_left","_right"))
df3
```
