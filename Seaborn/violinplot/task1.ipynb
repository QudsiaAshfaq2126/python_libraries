```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x="day",y="total_bill",kind="violin",data=tips)
```

```python
sns.catplot(x="day",y="total_bill",kind="violin",data=tips,hue="sex")
```

```python
sns.catplot(x="day",y="total_bill",kind="violin",data=tips,hue="sex",split="True")
```
