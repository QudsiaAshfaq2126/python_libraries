```python
import seaborn as sns
import numpy as np
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.catplot(x="smoker",y="tip",kind="bar",data=tips)
# black line shoe mean of the population with 95 confident interval
```

```python
sns.catplot(x="smoker",y="tip",kind="bar",data=tips,hue="sex")
```

```python
sns.catplot(x="smoker",y="tip",kind="bar",data=tips,hue="sex",estimator=np.median)
```

```python
sns.catplot(x="sex",kind="count",data=tips,hue="smoker")
```
