```python
import seaborn as sns
# import numpy as sns
```

```python
flights=sns.load_dataset("flights")
flights.head()
```

```python
x=flights.pivot_table(index="year",columns="month",values="passengers",aggfunc="sum")
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap="winter") # cmap sa color change hoty
```

# cluster map

```python
sns.clustermap(x)
```
