```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
revenue=[23,45,12,67]
company=["google","market","microsoft","F8"]
profit=[23,27,18,53]
```

```python
xpos=np.arange(len(company))
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
```

```python
plt.bar(xpos-0.2,revenue,width=0.2,label="revenue")
plt.bar(xpos+0.2,profit,width=0.2,label="profit")
plt.ylabel("revenue")
plt.title("UK stock")
plt.legend()
```

```python
plt.barh(xpos-0.2,revenue,label="revenue")
plt.barh(xpos+0.2,profit,label="profit")
plt.ylabel("revenue")
plt.title("UK stock")
plt.legend()
```
