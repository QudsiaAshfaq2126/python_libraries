```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[12,13,14,15,16,17,18]
min_t=[78,56,43,23,78,45,90]
avg_t=[56,34,67,23,78,12,56]
```

```python
plt.plot(days,min_t)
plt.plot(days,max_t)
plt.plot(days,avg_t)
```

# Label

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")
plt.plot(days,min_t)
plt.plot(days,max_t)
plt.plot(days,avg_t)
```

# Legends

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")
plt.plot(days,min_t,label="Min")
plt.plot(days,max_t,label="max")
plt.plot(days,avg_t,label="avg")
plt.legend(shadow=True ,fontsize="small") #legend set kr skty upper right,uper left,best
```

# Grid

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")
plt.plot(days,min_t,label="Min")
plt.plot(days,max_t,label="max")
plt.plot(days,avg_t,label="avg")
plt.legend(shadow=True ,fontsize="small")
plt.grid()
```
