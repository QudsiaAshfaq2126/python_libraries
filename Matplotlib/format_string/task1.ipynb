```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=[4,6,7,8,3,2]
y=[45,67,23,45,89,12]
```

```python
plt.plot(x,y,"g+")
```

```python
plt.plot(x,y,color="green",marker="+",linestyle=" ",markersize=3)
```

```python
plt.plot(x,y,color="green",marker=".",linestyle="--",markersize=3,alpha=0.5) # alpha show the apperance mean colour dim or bright
# alpha values between 0 and 1
```
