```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sugar_level=[113,85,90,150,159,88,93,115,135,80,77,82,129]
plt.hist(sugar_level,bins=3,rwidth=0.75) 
# bydefault 10 bins
```

```python
sugar_level=[113,85,90,150,159,88,93,115,135,80,77,82,129]

plt.hist(sugar_level,bins=[80,100,125,150],rwidth=0.95,color="r",histtype="step")
```

```python
sugar_level=[113,85,90,150,159,88,93,115,135,80,77,82,129]
sugar_level_women=[80,90,100,110,120,113,114,140,150,78,120,130,140]
plt.hist([sugar_level,sugar_level_women],bins=[80,100,125,150],rwidth=0.95,color=["r","g"],label=["man","women"])
plt.legend()
```

```python
sugar_level=[113,85,90,150,159,88,93,115,135,80,77,82,129]
sugar_level_women=[80,90,100,110,120,113,114,140,150,78,120,130,140]
plt.hist([sugar_level,sugar_level_women],bins=[80,100,125,150],rwidth=0.95,color=["r","g"],label=["man","women"],orientation="horizontal")
plt.legend()
```
